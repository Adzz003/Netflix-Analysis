# Which month has the highest rating on netflix?

```python
plt.figure(figsize = (20,20))
sns.countplot(y='month',data = df,hue='rating')
plt.title("Month vs Rating chart",size=18)
```

```python
df['release_year'] = df['release_year'].apply(str)
```

```python
plt.figure(figsize=(12,6))
df[df["type"]=="Movie"]["release_year"].value_counts()[:10].plot(kind="barh",color="red")
df[df["type"]=="TV Show"]["release_year"].value_counts()[:10].plot(kind="barh",color="green")
plt.title("TV shows and movies which were released in different years",size=18)
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x=df.director,order=df.director.value_counts().index[0:10])
plt.xticks(rotation=45)
plt.show()
```
